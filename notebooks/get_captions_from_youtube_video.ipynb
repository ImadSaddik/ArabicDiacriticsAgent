## Extract captions

Start by providing the video URL.

```python
from pytubefix import YouTube

video_url = "https://www.youtube.com/watch?v=E9-k7wtS3bg"
youtube = YouTube(video_url)
print(youtube.title)
```

Then, look at the list of available captions.

```python
captions = youtube.captions
captions
```

```python
arabic_captions = captions['a.ar']
captions_text = arabic_captions.generate_srt_captions()
print(captions_text)
```
